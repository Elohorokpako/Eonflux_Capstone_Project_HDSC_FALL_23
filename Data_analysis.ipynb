#### Due to the large size of the data, we will be importing from a the Eonflux folder in the drive

```python
# The base Google Drive link
base_link = 'https://drive.google.com/uc?id='

# The file ids from the shareable links
file_ids = ['1mOTTkRpx4W96D8zyTSqshpz46RRGDwiP', 
            '12Z6UcKk2vJJ2SrjldkHCpv41WR7XwEdI', 
            '1oo231pvNcbMz_biJB7V499H2Wt6FiGll', 
            '1MN7LitDc_AnPtPYpDLLVH5yqg357iFcg', 
            '1u1uzTFu3Z2PGkGwaDO5y5mj99YYdsm51']  # Replace with your file ids

# Create a dictionary to hold the dataframes
dfs = {}

# Loop over the list of file ids
for file_id in file_ids:
    # Combine the base link and file id
    file_link = base_link + file_id

    # Download the file
    output_file = f'{file_id}.csv'
    gdown.download(file_link, output_file, quiet=False)
    
    # Read the file into a DataFrame
    df = pd.read_csv(output_file)
    
    # Store the DataFrame in the dictionary
    dfs[file_id] = df
```
